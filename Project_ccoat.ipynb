# Stacking

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_predict, train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=100)
```

```python
model_1_list = [LogisticRegression(),DecisionTreeClassifier(),
                RandomForestClassifier(n_estimators=100),KNeighborsClassifier(n_neighbors=3),xgb.XGBClassifier()]
model_2_list = [LogisticRegression(),DecisionTreeClassifier(),
                RandomForestClassifier(n_estimators=100),KNeighborsClassifier(n_neighbors=3),
                SVC(kernel='linear'),xgb.XGBClassifier()]

output_list = []


for i in model_1_list:
    model_1 = i
    model_2 = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)

    # Use cross-validation to generate meta-features
    m1_meta_features = cross_val_predict(model_1, X_train, y_train, cv=5, method='predict_proba')[:, 1]
    m2_meta_features = cross_val_predict(model_2, X_train, y_train, cv=5, method='predict_proba')[:, 1]
    meta_features = pd.DataFrame({'Model_1': m1_meta_features, 'Gradient Boosting': m2_meta_features})

    # Train the base models on the full training set
    model_1.fit(X_train, y_train)
    model_2.fit(X_train, y_train)

    # Generate meta-features for the test set
    m1_test_meta_features = model_1.predict_proba(X_test)[:, 1]
    m2_test_meta_features = model_2.predict_proba(X_test)[:, 1]
    test_meta_features = pd.DataFrame({'Model_1': m1_test_meta_features, 'Gradient Boosting': m2_test_meta_features})

    for j in model_2_list:
        meta_model = j
        meta_model.fit(meta_features, y_train)
        y_pred = meta_model.predict(test_meta_features)
        #print('Accuracy: {:.2f}%'.format(accuracy_score(y_test, y_pred)*100))
        output = '{:.2f}%'.format(accuracy_score(y_test, y_pred)*100)
        output_list.append(output)

dff1 = pd.DataFrame([output_list[i:i+6] for i in range(0, 30, 6)],
                  index=['Logistic','DecisionTree',
                                 'RandomForest','KNeighbors','XGBoost'],
                  columns=['Logistic','DecisionTree',
                                   'RandomForest','KNeighbors','SVM','XGBoost'])
print(dff1)
```

# Ensembling using Voting Classifier

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict, train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, 
                                                    test_size=0.2, random_state=100)
```

```python
models = [LogisticRegression(),DecisionTreeClassifier(),
          RandomForestClassifier(),KNeighborsClassifier(n_neighbors=3),
          SVC(kernel='linear'),xgb.XGBClassifier()]

output_list = []
    
for i in models:
    model_1 = i
    for j in models:
        model_2 = j
        model_3 = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=100)

        ensemble_model = VotingClassifier(estimators=[('m1', model_1),('m2', model_2),('m3', model_3)], voting='hard')
        ensemble_accuracy = cross_val_predict(ensemble_model, X_train, y_train, cv=5)
        ensemble_model.fit(X_train, y_train)
        y_pred = ensemble_model.predict(X_test)
        
        
        output = '{:.2f}%'.format(accuracy_score(y_test, y_pred)*100)
        output_list.append(output)

dff2 = pd.DataFrame([output_list[i:i+6] for i in range(0, 36, 6)],
                  index=['Logistic','DecisionTree','RandomForest','KNeighbors','SVM','XGBoost'],
                  columns=['Logistic','DecisionTree','RandomForest','KNeighbors','SVM','XGBoost'])
print(dff2)
        
```

# Bagging

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=100)
```

```python
models = [LogisticRegression(),DecisionTreeClassifier(),
          RandomForestClassifier(),KNeighborsClassifier(n_neighbors=3),
          SVC(kernel='linear'),xgb.XGBClassifier()]

output_list = []

for i in models:
    tree=i
    
    # define the bagging classifier with 10 estimators
    bagging = BaggingClassifier(estimator=tree, n_estimators=10)

    # train the bagging classifier on the training data
    bagging.fit(X_train, y_train)

    # evaluate the bagging classifier on the test data
    accuracy = bagging.score(X_test, y_test)
    accuracy = '{:.2f}%'.format(accuracy*100)
    output_list.append(accuracy)

dff3 = pd.DataFrame([output_list],index=['Accuracy'],
                 columns=['Logistic','DecisionTree','RandomForest','KNeighbors','SVM','XGBoost'])
print(dff3)
```

# Majority Voting

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=100)
```

```python
models = [LogisticRegression(),DecisionTreeClassifier(),
          RandomForestClassifier(),KNeighborsClassifier(n_neighbors=3),
          SVC(kernel='linear'),xgb.XGBClassifier()]

output_list = []
    
for i in models:
    model_1 = i
    for j in models:  
        model_2 = VotingClassifier(estimators=[('m1', model_1), ('m2', model_2), ('m3', model_3)], voting='hard')
        model_3 = j
        
        # Train each base model on the full training set
        model_1.fit(X_train, y_train)
        model_2.fit(X_train, y_train)
        model_3.fit(X_train, y_train)
        
        pred_1 = model_1.predict(X_test)
        pred_2 = model_2.predict(X_test)
        pred_3 = model_3.predict(X_test)
        
        # Combine the predictions using majority voting
        voting_pred = (pred_1 + pred_2 + pred_3) / 3
        voting_pred = [1 if x > 0.5 else 0 for x in voting_pred]

        output = '{:.2f}%'.format(accuracy_score(y_test, voting_pred)*100)
        output_list.append(output)

dff4 = pd.DataFrame([output_list[i:i+6] for i in range(0, 36, 6)],
                  index=['Logistic','DecisionTree','RandomForest','KNeighbors','SVM','XGBoost'],
                  columns=['Logistic','DecisionTree','RandomForest','KNeighbors','SVM','XGBoost'])
print(dff4)
```
